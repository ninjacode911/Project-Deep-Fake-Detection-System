```python
%%writefile /content/Efficientmodel/fine_tune.py
import os
import logging
import torch
import torch.nn as nn
import pandas as pd
import ast
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader
import cv2
import timm
import albumentations as A
from albumentations.pytorch import ToTensorV2

# Set up logging
logging.basicConfig(level=logging.INFO, filename='/content/drive/MyDrive/DeepFake/logs/finetune.log', filemode='a',
                    format='%(asctime)s - %(levelname)s - %(message)s')

# Constants
PROCESSED_DATA_PATH = '/content/processed'
PRETRAINED_MODEL_PATH = '/content/drive/MyDrive/DeepFake/pretrained/efficientnetv2_s_pretrained.pth'
BATCH_SIZE = 16
NUM_EPOCHS = 10
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Data augmentation and normalization
transform = A.Compose([
    A.Resize(224, 224),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2()
])

# Custom Dataset
class DeepFakeDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        file_path = self.dataframe.iloc[idx]['file_path']
        label = self.dataframe.iloc[idx]['label']

        if file_path.endswith('.npy'):  # Spectrogram
            spec = np.load(file_path)
            spec = (spec - spec.min()) / (spec.max() - spec.min() + 1e-6)
            plt.figure(figsize=(224/100, 224/100), dpi=100)
            plt.specgram(spec, Fs=2)
            plt.axis('off')
            plt.savefig('temp.png', bbox_inches='tight', pad_inches=0)
            plt.close()
            image = cv2.imread('temp.png')
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            os.remove('temp.png')
        else:  # Frame image
            image = cv2.imread(file_path)
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        if image is None:
            logging.error(f"Failed to load file: {file_path}")
            raise ValueError(f"Failed to load file: {file_path}")

        if self.transform:
            augmented = self.transform(image=image)
            image = augmented['image']

        return image, torch.tensor(label, dtype=torch.long)

# Load and preprocess data
datasets = ['AVLips']
all_metadata = []

for dataset in datasets:
    metadata_path = os.path.join(PROCESSED_DATA_PATH, dataset, 'metadata.csv')
    if not os.path.exists(metadata_path):
        logging.warning(f"Metadata file not found for dataset {dataset}: {metadata_path}")
        continue

    logging.info(f"Processing dataset: {dataset}")
    df = pd.read_csv(metadata_path)
    logging.info(f"Number of rows in {dataset}/metadata.csv: {len(df)}")

    unique_labels = df['label'].unique()
    logging.info(f"Unique labels in {dataset}: {unique_labels}")

    df = df.dropna(subset=['frames', 'spectrograms'])

    # Process frames
    df_frames = df.copy()
    try:
        df_frames['frames'] = df_frames['frames'].apply(ast.literal_eval)
    except ValueError as e:
        logging.error(f"Error parsing 'frames' in {dataset}: {e}")
        continue
    df_frames_exploded = df_frames.explode('frames').reset_index(drop=True)
    df_frames_exploded = df_frames_exploded.drop(columns=['file_path'], errors='ignore')
    df_frames_exploded = df_frames_exploded.rename(columns={'frames': 'file_path'})
    df_frames_exploded['file_path'] = df_frames_exploded['file_path'].apply(
        lambda x: x.replace('HDD/Datasets/processed', '/content/processed')
    )
    logging.info(f"Number of frames in {dataset}: {len(df_frames_exploded)}")
    logging.info(f"df_frames_exploded columns: {list(df_frames_exploded.columns)}")

    # Process spectrograms
    df_spectrograms = df.copy()
    try:
        df_spectrograms['spectrograms'] = df_spectrograms['spectrograms'].apply(ast.literal_eval)
    except ValueError as e:
        logging.error(f"Error parsing 'spectrograms' in {dataset}: {e}")
        continue
    df_spectrograms_exploded = df_spectrograms.explode('spectrograms').reset_index(drop=True)
    df_spectrograms_exploded = df_spectrograms_exploded.drop(columns=['file_path'], errors='ignore')
    df_spectrograms_exploded = df_spectrograms_exploded.rename(columns={'spectrograms': 'file_path'})
    df_spectrograms_exploded['file_path'] = df_spectrograms_exploded['file_path'].apply(
        lambda x: x.replace('HDD/Datasets/processed', '/content/processed')
    )
    logging.info(f"Number of spectrograms in {dataset}: {len(df_spectrograms_exploded)}")
    logging.info(f"df_spectrograms_exploded columns: {list(df_spectrograms_exploded.columns)}")

    # Combine frames and spectrograms, selecting only 'file_path' and 'label'
    try:
        df_combined = pd.concat([
            df_frames_exploded[['file_path', 'label']],
            df_spectrograms_exploded[['file_path', 'label']]
        ], ignore_index=True)
    except Exception as e:
        logging.error(f"Error during pd.concat in {dataset}: {e}")
        raise

    if len(df_combined) > 0:
        sample_path = df_combined['file_path'].iloc[0]
        logging.info(f"Sample file path in {dataset}: {sample_path}")
        if not os.path.exists(sample_path):
            logging.error(f"Sample file does not exist: {sample_path}")

    # Map labels
    df_combined['label'] = df_combined['label'].str.lower()
    label_mapping = {'likely fake': 0, 'fake': 1, 'neutral': 2, 'real': 3, 'likely real': 4}
    df_combined['label'] = df_combined['label'].map(label_mapping)

    unmapped_labels = df_combined['label'].isna().sum()
    if unmapped_labels > 0:
        logging.warning(f"Found {unmapped_labels} rows with unmapped labels in {dataset}")

    df_combined = df_combined.dropna(subset=['file_path', 'label'])
    logging.info(f"Number of entries in {dataset} after preprocessing: {len(df_combined)}")

    if len(df_combined) > 0:
        all_metadata.append(df_combined)

if not all_metadata:
    raise ValueError("No data loaded from any dataset. Check metadata and file paths.")

metadata = pd.concat(all_metadata, ignore_index=True)
logging.info(f"Total metadata rows: {len(metadata)}")

if len(metadata) == 0:
    raise ValueError("No data loaded after preprocessing. Check metadata and file paths.")

# Split data
train_df, val_df = train_test_split(metadata, test_size=0.2, random_state=42)
train_dataset = DeepFakeDataset(train_df, transform=transform)
val_dataset = DeepFakeDataset(val_df, transform=transform)

if len(train_dataset) == 0:
    raise ValueError("Training dataset is empty. Check metadata and file paths.")

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)

logging.info(f"Train dataset size: {len(train_dataset)}")
logging.info(f"Validation dataset size: {len(val_dataset)}")

# Load model
try:
    model = timm.create_model('efficientnetv2_s', pretrained=True, num_classes=5)
    model.load_state_dict(torch.load(PRETRAINED_MODEL_PATH), strict=False)
except Exception as e:
    logging.error(f"Error loading pretrained model: {e}")
    raise
model = model.to(DEVICE)

# Training setup
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
best_val_acc = 0.0
checkpoint_path = "/content/drive/MyDrive/DeepFake/pretrained/efficientnetv2_s_finetuned.pth"

# Training loop
for epoch in range(NUM_EPOCHS):
    model.train()
    train_loss, train_correct, train_total = 0, 0, 0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()
        _, predicted = outputs.max(1)
        train_total += labels.size(0)
        train_correct += predicted.eq(labels).sum().item()

    train_acc = 100. * train_correct / train_total
    logging.info(f"Epoch {epoch+1}/{NUM_EPOCHS}, Train Loss: {train_loss/len(train_loader):.4f}, Train Acc: {train_acc:.2f}%")

    # Validation
    model.eval()
    val_loss, val_correct, val_total = 0, 0, 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            val_loss += loss.item()
            _, predicted = outputs.max(1)
            val_total += labels.size(0)
            val_correct += predicted.eq(labels).sum().item()

    val_acc = 100. * val_correct / val_total
    logging.info(f"Epoch {epoch+1}/{NUM_EPOCHS}, Val Loss: {val_loss/len(val_loader):.4f}, Val Acc: {val_acc:.2f}%")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), checkpoint_path)
        logging.info(f"Saved best model with Val Acc: {best_val_acc:.2f}%")

logging.info("Fine-tuning completed")
```
